```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
df = pd.read_csv("master.csv")
```

```python
male_rate_mean = df[df['sex']=='male']['suicides/100k pop'].mean()
female_rate_mean = df[df['sex']=='female']['suicides/100k pop'].mean()
print(male_rate_mean,female_rate_mean)
# average male vs female suicide rate over all years and country
# male suicide rate is significantly higher than female 
```

```python
male_ratee = df[df['sex']=='male']['suicides/100k pop']
female_ratee = df[df['sex']=='female']['suicides/100k pop']
plot = plt.boxplot([male_ratee,female_ratee])
plt.title('Suicide by Gender')
plt.ylabel("suicides/100k pop")
plt.xticks([1,2],['male','female'])
# shows lower average and spread 
```

```python
from statsmodels.graphics.gofplots import qqplot_2samples
import statsmodels.api as sm
pp_x = sm.ProbPlot(male_ratee)
pp_y = sm.ProbPlot(female_ratee)
qqplot_2samples(pp_x, pp_y, line='45')
plt.title('Suicide by Gender')
plt.ylabel("male")
plt.xlabel("female")
plt.show()
# y values are lower than x values
# male suicides rates are higher than females 
```

```python
df[df['sex']=='male']['suicides/100k pop'].mean()
Generations = ["Generation X", "Silent", "Millenials", "Boomers", "G.I. Generation"]
#1901 to 1927
gi_gen = df[df['generation']=='G.I. Generation']['suicides/100k pop']
#1928 – 1945
silent = df[df['generation']=='Silent']['suicides/100k pop']
#1946 – 1964
boomers = df[df['generation']=='Boomers']['suicides/100k pop']
#1965 – 1980
gen_x = df[df['generation']=='Generation X']['suicides/100k pop']
#1981-1996
millen = df[df['generation']=='Millenials']['suicides/100k pop']

plot = plt.boxplot([gi_gen,silent,boomers,gen_x,millen])
# Group by generations. 
# Doesn't tell much because younger age has less suicide numbers, 
# but can increase over the years
```

```python
year = []
suicide_mean = []

for i in range(1985,2016):
    year.append(i)
    tempt = df[df['year']==i]
    mean = tempt[tempt['sex']=='female']['suicides/100k pop'].mean()
    suicide_mean.append(mean)
    
plt.scatter(year, suicide_mean)
plt.title('Female suicide by year')
plt.ylabel("average female suicide/100pop")
plt.xlabel("year")
```
